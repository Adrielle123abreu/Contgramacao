# Arquivos txt

**txt --> Um arquivo de texto (ou arquivo .txt) é um formato de arquivo simples que armazena dados de texto, sem formatação especial, como fontes, tamanhos ou cores.**

1.   Documento de texto





# Abrindo arquivos txt


Usa-se **read** para ler arquivos

* Lê o conteúdo todo como se fosse uma string


* Ao abrir um arquivo, apesar de não aparecer no computador, ele permance aberto no python até ser fechado.








---



```python
Cont = open("Contgramação.txt", "r")
print(Cont.read())
Cont.close()
```

# Método With



* Utilizando o **with**, assim que o arquivo for aberto e rodado o código, ele automaticamente é fechado.









```python
with open("Contgramação.txt", "r") as arquivo:
  Cont = arquivo.read()
print(Cont)
```

```python
with open("Alunos.txt", "r") as arquivo:
  alunos = arquivo.readlines()
print(alunos)
```

```python
for hashtag_yt_org in alunos:
    print(hashtag_yt_org)
```

Usa-se **readlines** para arquivos maiores



* Lê todas as linhas do arquivo e retorna uma lista de strings, onde cada string representa uma linha do arquivo.
*   É criado uma lista onde cada item é uma linha do texto.



---



```python
with open("Contgramação.txt","r") as arquivo2:
  Alunos = arquivo2.readlines()
Alunos
```

```python
for linha in Alunos:
  if "Arthur" in linha:
    print("Líder:",linha)
  elif "Natalia" in linha:
    print("Presidente:", linha)
  elif "Ana" in linha:
    print("Vice:",linha)
  else:
    print("Membro:",linha)
```

# Escrevendo no arquivo

* W --> write

* Segue a mesma lógica do read, apenas é preciso mudar o **r** por **w**

* Se o arquivo não existe é criado um, e se já existir será substituido toda a informação do existente.   









```python
with open("Contgramação.txt","w") as arquivo:
 arquivo.write("Leticia Amy Siramidu")
```

```python
with open("DEV.txt","w") as arquivo:
 arquivo.write("Melhor entidade da USP")
```

# Alterando o arquivo

* a --> append




```python
with open("Contgramação2.txt","a") as arquivo:
 arquivo.write("\nLeticia Amy Siramidu")
```
